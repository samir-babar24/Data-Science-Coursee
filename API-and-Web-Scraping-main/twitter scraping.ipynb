```python
#now simply creating function to have ease in takin tweeter lins, mode, and no of tweets
#creating dataframe for extracted specific data

import pandas as pd
from ntscraper import Nitter

def easyfunbysamir(name,modes,no):
    scraper=Nitter()#used to remove risk and privatetization # this line needs to be included in the function
    raj  = scraper.get_tweets(name,mode=modes,number=no)
    #creating an dummy array
    #dummy
    samirbabar=[]
    for tweets in raj['tweets']:
        data=(tweets['link'],tweets['text'],tweets['date'])
        samirbabar.append(data)
    vijaydhame=pd.DataFrame(samirbabar,columns=['link','text','Date'])
    return vijaydhame
```

```python
easyfunbysamir('callofduty','term',20)
```
