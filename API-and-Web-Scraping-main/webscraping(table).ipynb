```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import html5lib

url="https://en.wikipedia.org/wiki/World_population"#importing url
cont=requests.get(url)#reading
samir=BeautifulSoup(cont.content,"html.parser")#parsing the distribute data
table1=samir.find('table',{"class":"wikitable sortable"})#finding the exact table location by using class,we can also do it by using id
#here wikipedia sortable is class of table
#print(table1) #herewe ce will get html code through
headers=table1.find_all("th")#here th means table heads(columnnames)
print(headers)




```

```python
#create dummy title array to save headers files data
titles=[]
for x in headers:
  title=x.text
  titles.append(title)#saving in a dummay array created over
print(titles)
```

```python
#findinf data in arrays and using a dummur array
rows=table1.find_all('tr')#here tr means table rows
data=[]
for x in rows:
  row_data=x.find_all('td')#now we know rows of the table but dont kow the exact data in a tble so td is tabels data
  #print(data)
  # now here we are getting very disributed data so we are trying to do print data's element as text
  neetdata=[tr.text for tr in row_data]
  data.append(neetdata)#saving in dummy array created over
  print(neetdata)
```

```python
#now creating a dataframe
raju=pd.DataFrame(data,columns=titles)
raju
```
