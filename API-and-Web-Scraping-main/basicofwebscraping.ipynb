```python
#importing for content ,parsing and opoeration on content
import requests
import html5lib
from bs4 import BeautifulSoup
#url of website

url="https://en.wikipedia.org/wiki/List_of_equipment_of_the_Indian_Army"
#get content
r=requests.get(url)
#html content
htmlcontent=r.content

#parse
soup=BeautifulSoup(htmlcontent,'html.parser')
print(soup.prettify)



```

```python
#html traversing the tree
#title of the web
tit=soup.title
print(tit)
```

```python
#paragraph
para=soup.find_all('p')
print(para)
```

```python
#finding anchor tags
anc=soup.find_all('a')
print(anc)
#for getting the links

```

```python
#to get first para info
fp=soup.find('p')
print(fp)
```

```python
#to find specific paras with class
spec_p=soup.find_all('p',{'class':'id'})#here id is one class
print(spec_p)
#here showing output as null means therer is no para with classname id
```

```python
#getting exact links from anchors
for link in anc:
 print(link.get('href'))
```

```python
#specific content in a id
id_read=soup.find(id="cite_ref-PIB_India_269-0")
print(id_read)
#childern in content of id(child tags)
id_read=soup.find(id="cite_ref-PIB_India_269-0")
print(id_read.children)

```

```python
#using for loop (u can do it for children,content text etx)
id_read=soup.find(id="cite_ref-PIB_India_269-0")
for elem in id_read.children:
  print(elem)

```
